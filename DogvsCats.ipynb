```python
import os
from PIL import Image

image_exts = ['jpeg', 'jpg', 'png']
data_dir = r"C:\Users\anshu\Desktop\DogsVSCats"

for root, dirs, files in os.walk(data_dir):
    for file in files:
        file_path = os.path.join(root, file)
        ext = os.path.splitext(file)[1].lower().strip('.')   # get extension without "."

        # Skip obvious non-images (like .ipynb, .txt, etc.)
        if ext not in image_exts:
            # optional: print to see what was skipped
            print(f"Skipping non-image file: {file_path}")
            continue

        try:
            with Image.open(file_path) as img:
                fmt = img.format.lower()
                if fmt not in image_exts:
                    print(f"Invalid format: {file_path} (detected {fmt})")
                    os.remove(file_path)
                else:
                    _ = img.copy()   # your processing

        except Exception as e:
            print(f"Corrupt or unreadable image {file_path}. Error: {e}")
            os.remove(file_path)
```

```python
import numpy as np
from PIL import Image
import os

def image_to_array(image_path):
    """
    Read an image and convert it to a numpy array.

    Parameters:
    - image_path (str): The path to the image file.

    Returns:
    - np.ndarray: The image as a numpy array, or None if reading failed.
    """
    if not isinstance(image_path, str):
        raise TypeError("image_path must be a string path to the image file.")
    if not os.path.exists(image_path):
        print(f"Could not read the image file at {image_path}. File does not exist.")
        return None

    try:
        with Image.open(image_path) as img:
            # Optionally convert to RGB to ensure consistent shape for single-channel images
            img = img.convert("RGB")
            image_array = np.array(img)
            return image_array
    except (IOError, OSError) as e:
        print(f"Could not read the image file at {image_path}. Error: {e}")
        return None
```

```python
base_dir = r'C:\Users\anshu\Desktop\DogsVSCats\train'
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
```

```python
import os
print(os.listdir(base_dir))
```

```python
train_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150), 
    batch_size=20,
    class_mode='binary',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150),  
    batch_size=20,
    class_mode='binary',     
    subset='validation'
)
```

```python
images, labels = next(train_generator)   # standardization/ rescale (1./255)
images, labels
```

```python
# accessing train data
train_class_labels = train_generator.class_indices 
print("Training class labels:", train_class_labels)

#accessing validation data
validation_class_labels = validation_generator.class_indices
print("validation class labels:", validation_class_labels)
```

```python
def plot_images_from_generator(generator, num_images=4):
    images, labels = next(generator)

    fig, axes = plt.subplots(1, num_images, figsize=(20,20))
    axes = axes.flatten()

    for img, label, ax in zip(images, labels, axes):
        ax.imshow(img)
        class_labels = dict((v,k) for k,v in generator.class_indices.items())
        ax.set_title(class_labels[int(label)])
        ax.axis('off')
    plt.tight_layout()
    plt.show()

plot_images_from_generator(train_generator, num_images=4)
        
```

```python
train_generator.reset()
```

```python
# CNN model #
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary() #all these conv2D n maxpooling are things going on known as CNN part 
```

```python
plot_model(model, to_file = 'Architecture.png',
           show_shapes=True, show_layer_names=True)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
batch = 20
train_steps = np.ceil(train_generator.samples / batch)
validation_steps_per_epoch = np.ceil(validation_generator.samples / batch)
print(train_steps, validation_steps_per_epoch)
```

```python
history = model.fit(
                    train_generator,
                    steps_per_epoch=800,
                    epochs=10,
                    validation_data=validation_generator,
                    validation_steps=200)
```

```python
history.history
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Reset the test generator
test_generator.reset()

# Get the true labels
true_classes = test_generator.classes

# Predict probabilities for the positive class
probabilities = model.predict(test_generator)  # <-- cleaner

# Convert probabilities to binary predictions
predicted_classes = (probabilities > 0.5).astype(int).squeeze()

# Get class labels (['cats','dogs'])
class_labels = list(test_generator.class_indices.keys())

predicted_classes
```

```python
# Generate the confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)

# Plotting with seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
def make_prediction(image_path):
    img = load_img(img_path, target_size=(150, 150))  # Resizing the image
    img_array = img_to_array(img)  # Converting to array
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array /= 255.  # Rescale the image similarly as we did for the training data

    prediction = model.predict(img_array)
    print("Prediction:", prediction)

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')

    # Prediction
    plt.subplot(1, 2, 2)
    if prediction[0][0] > 0.5:
        plt.title('Predicted: Dog')
    else:
        plt.title('Predicted: Cat')
    plt.imshow(img)  # Show the same image as prediction
    plt.axis('off')

    plt.show()
```
